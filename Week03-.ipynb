## Mising Value

```python
df.info()
#untuk mengetahui info dari variabel df
```

```python
jumlah_missing = pd.isnull(df).sum()
# untuk mengetahui data yang kosong ada berapa dan kemudian kita inisialisasikan ke variabel
```

```python
rasio = jumlah_missing.div(len(df))
rasio
#untuk mengetahui berapa rasion dari data missing
```

```python
df = df.drop(['ADDRESSLINE2'], axis=1)
# untuk mengahapus ADDRESSLINE2 
```

```python
#untuk mengecek variabel df
df.head()
```

```python
df['QUANTITYORDERED'].describe()
#untuk mencari penjelasan/ ringkasan dari QUANTITYORDERED
```

```python
df['QUANTITYORDERED'].fillna(int(df['QUANTITYORDERED'].mean()), inplace=True)
#mengisi data yang kosog
```

```python
#untuk mengecek kembali data yang kosong
pd.isnull(df).sum()
```

```python
df[df[['ORDERNUMBER', 'ORDERLINENUMBER']].duplicated(keep=False)].sort_values('ORDERNUMBER')
#Mencari data yang sama/ duplikat
```

```python
#kita update data df nya dan kita reset index nya karna ada data yang duplikat
df = df.drop_duplicates(subset=['ORDERNUMBER','ORDERLINENUMBER']).reset_index()
```

```python
df[df[['ORDERNUMBER', 'ORDERLINENUMBER']].duplicated(keep=False)].sort_values('ORDERNUMBER')
#Mencari data yang sama/ duplikat lagi apakah data nya masih duplikat atau sudah di merge
```

```python
df['QUANTITYORDERED'].dtypes
#mencari tipe data dari variabel
```

```python
#kita ubdah tipe data dari float ke int dan kita inisialisasikan ke variabel baru
df['QUANTITYORDERED_int'] = df['QUANTITYORDERED'].astype('Int64')
```

```python
df.info()
#untuk mengetahui info dari variabel df
```

```python
#untuk mengubah nama variabel dalam dataset
df = df.rename(columns={'ORDERNUMBER' : 'Nomor_pesanan', 'PRICEEACH' : 'harga_satuan',
                         'QUANTITYORDERED_int' : 'Quantity'})
```

```python
df.info()
#untuk mengetahui info dari variabel df
```

mengambil data data yang kita butuhkan saja

**Grouping**
Untuk meenggabungkan variabel yang sama yang ada di quality dan sales

```python
quality = df_fil_clean[['Nomor_pesanan', 'QUANTITYORDERED']]
sales = df_fil_clean[['Nomor_pesanan', 'SALES']]

quality.head()
```

```python
df_new = quality.merge(sales, how='left', on='Nomor_pesanan')
df_new.head()
```

```python
harga_satu = df_fil_clean[['Nomor_pesanan', 'harga_satuan']]
status = df_fil_clean[['Nomor_pesanan', 'STATUS']]

harga_satu.head()
```

```python
df_new2 = harga_satu.merge(status, how='left', on='Nomor_pesanan')
df_new2.head()
```

**Group by**

Fungsi GroupBy() memungkinkan kita untuk mengelompokkan data dalam kumpulan item yang sama misalnya dalam status, no pesanan,dan kategori lainnya

```python
df_fil_clean.groupby('STATUS')['Nomor_pesanan'].count()
```
