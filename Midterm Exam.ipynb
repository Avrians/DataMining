```python
#menampilkan data yang berbeda dalam kolom tertentu
items =(data['4'].unique())
items
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# penggantian data
# melakukan encoding
itemset = set(items)
encoded_vals = []
for index, row in data.iterrows():
  rowset = set(row)
  labels = {}
  uncommons = list(itemset - rowset)
  commons = list(itemset.intersection(rowset))
  for uc in uncommons:
    labels[uc] = 0
  for com in commons:
    labels[com] = 1
  encoded_vals.append(labels)
encoded_vals[0]
ohe_data = pd.DataFrame(encoded_vals)
```

```python
ohe_data
```

```python
data_retail = ohe_data.drop(columns=ohe_data.columns[1])
data_retail
```

```python
data_retail.info()
```

```python
baris_data = data_retail.iloc[0:101, [7,0,4,5]]
baris_data
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# baru kita membuat sesuai dengan apa yang ditentukan
# misal kita ingin minum support nya 0,2
freq_items = apriori(baris_data, min_support=0.2, use_colnames=True)
freq_items
```

```python
rules = association_rules(freq_items, metric="confidence", min_threshold=0.3)
rules
```

```python
freq_items_2 = apriori(baris_data, min_support=0.1, use_colnames=True)
freq_items_2
```

```python
rules_2 = association_rules(freq_items_2, metric="confidence", min_threshold=0.3)
rules_2
```
